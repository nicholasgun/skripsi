```python
import pandas as pd
```

```python
# get the issues (read json file)
issues = pd.read_json('../../Data/Data From API/Kind:Bug/github_bug_issues.json')

# get the shape of the dataframe
print(issues.shape)
```

```python
issues.head(5)
```

```python
# get the link between pr and issues
links = pd.read_json('../../Data/Data From API/Kind:Bug/pr_bug_links_new.json')

print(links.shape)
```

```python
links.head(5)
```

```python
# merge the issues and links
data = issues.merge(links, left_on='html_url', right_on='issue_url', how='inner')

# remove html_url column
data.drop(columns=['html_url'], inplace=True)

print(f'new shape: {data.shape}')
```

```python
data.head(5)
```

```python
# leave the comments_url, id, title, body, issue_url, and pr_url columns
data = data[['comments_url', 'id', 'title', 'body', 'issue_url', 'pr_url']]
data.head(5)
```

```python
# get the pr (read json file)
prs = pd.read_json('../../Data/Data From API/Kind:Bug/merged_pull_requests_with_label_kind_bug_2015_until_2025.json')
print(prs.shape)
```

```python
prs.head(5)
```

```python
# only leaves the html_url, labels, and pr number
prs = prs[['html_url', 'labels', 'number']]
prs.head(5)
```

```python
# merge the prs and the data
data = data.merge(prs, left_on='pr_url', right_on='html_url', how='inner')

# remove html_url column
data.drop(columns=['html_url'], inplace=True)

print(f'new shape: {data.shape}')
```

```python
data.head(5)
```

```python
# extract the labels name only from labels column
data['labels'] = data['labels'].apply(lambda x: [label['name'] for label in x])

# filter the labels name to only have the area labels
data['labels'] = data['labels'].apply(lambda x: [label for label in x if label.startswith('area')])
```

```python
print(f'before filter: {data.shape}')

# delete the row with no labels
data = data[data['labels'].apply(lambda x: len(x) > 0)]

print(f'after filter: {data.shape}')
```

```python
data.head(5)
```

```python
changed_files = pd.read_csv('../../Data/Data From API/Kind:Bug/pr_changed_files.csv')
changed_files.head(5)
```

```python
print(f'before merge: {data.shape}')

# merge the data with changed_files
data = data.merge(changed_files, left_on='number', right_on='pr_number', how='inner')

# remove pr_number column
data.drop(columns=['number'], inplace=True)

print(f'new shape: {data.shape}')
```

```python
data.head(5)
```

```python
# count the frequency of every label name
from collections import Counter

labels = Counter([label for labels in data['labels'] for label in labels])

for label, count in labels.most_common():
    print(f'{label}: {count}')
```

```python
files = Counter([
    file.strip(" '")
    for fn in data['filename']
    for file in fn.strip("[]").split(',')
])

for file, count in files.most_common():
    print(f'{file}: {count}')
```

```python
# export the data to json file
data.to_json('../../Data/Preprocessed Data/kind:bug/merged_data.json', orient='records')
```
