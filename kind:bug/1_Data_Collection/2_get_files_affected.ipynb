```python
import pandas as pd
import requests
import json
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
datas = pd.read_json('../../Data/Data From API/Kind:Bug/merged_pull_requests_with_label_kind_bug_2015_until_2025.json')
datas
```

```python
# Konfigurasi
BASE_URL = "https://api.github.com"
REPO_OWNER = "kubernetes"  # Ganti dengan nama owner repositori
REPO_NAME = "kubernetes"  # Ganti dengan nama repositori
TOKEN = None  # Masukkan token GitHub Anda
```

```python
datas
```

```python
def get_changed_files(pr_number):
    url = f"{BASE_URL}/repos/{REPO_OWNER}/{REPO_NAME}/pulls/{pr_number}/files"
    headers = {
        "Accept": "application/vnd.github.v3+json",
        "Authorization": f"token {TOKEN}"
    }
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            files = response.json()
            result = {
                "pr_number" : pr_number,
                "filename" : [file["filename"] for file in files], 
                "status" : [file["status"] for file in files],
                "additions" : [file["additions"] for file in files],
                "deletions" : [file["deletions"] for file in files],
                "changes" : [file["changes"] for file in files]
                }
            
            return result
        else:
            print(f"Error {response.status_code}: {url}")
            return None
    except Exception as e:
        print(f"Error fetching {url}: {str(e)}")
     
```

```python
def get_changed_files_parallel(pr_numbers):
    with ThreadPoolExecutor(max_workers=5) as executor:
        futures = [executor.submit(get_changed_files, pr_number) for pr_number in pr_numbers]
        results = []
        for future in tqdm(as_completed(futures), total=len(pr_numbers)):
            result = future.result()
            if result:
                results.append(result)
        return results
```

```python
# use parallel processing to get changed files faster
changed_files = get_changed_files_parallel(datas["number"])
```

```python
changed_files_df = pd.DataFrame(changed_files)
changed_files_df
```

```python
# export to csv
changed_files_df.to_csv("pr_changed_files.csv", index=False)
```
