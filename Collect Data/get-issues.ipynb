# Get all issues

```python
# Header untuk otentikasi
headers = {
    "Authorization": f"Bearer {TOKEN}",
    "Accept": "application/vnd.github.v3+json"
}

# Endpoint untuk mendapatkan issues
endpoint = f"/repos/{REPO_OWNER}/{REPO_NAME}/issues"

# Parameter (opsional)
params = {
    "state": "closed",  # Mengambil semua issues (open dan closed)
    "labels": "kind/feature", # Filter berdasarkan label kind/feature
    "per_page": 100,  # Maksimum jumlah item per halaman
    "page": 1        # Halaman pertama
}

# Variabel untuk menyimpan semua issues
all_issues = []

# Loop untuk menangani pagination
while True:
    # Permintaan ke API GitHub
    response = requests.get(f"{BASE_URL}{endpoint}", headers=headers, params=params)
    if response.status_code == 200:
        issues = response.json()
        if not issues:  # Jika tidak ada data lagi, hentikan loop
            break
        all_issues.extend(issues)  # Tambahkan issues ke dalam daftar
        params["page"] += 1  # Pindah ke halaman berikutnya
    else:
        print(f"Permintaan gagal dengan status: {response.status_code}")
        break

# Simpan semua issues ke dalam file JSON
with open("github_issues_3.json", "w") as file:
    json.dump(all_issues, file, indent=4)

print("Semua data issues berhasil disimpan.")
```

# Get Merged PR with feature label

```python
import requests
import json
from datetime import datetime, timedelta

# GitHub repository and authentication details
BASE_URL = "https://api.github.com/search/issues"

# Headers for the API request
HEADERS = {
    "Authorization": f"Bearer {TOKEN}",
    "Accept": "application/vnd.github.v3+json"
}

def search_merged_pull_requests_with_label(label, start_date, end_date):
    """
    Searches for merged pull requests with a specific label within a date range.
    :param label: The label to filter pull requests by
    :param start_date: Start date for filtering (YYYY-MM-DD)
    :param end_date: End date for filtering (YYYY-MM-DD)
    :return: List of merged PRs with the specified label in the date range
    """
    merged_prs = []
    page = 1

    while True:
        # Construct the search query with the label and date range filter
        query = (f"repo:{REPO_OWNER}/{REPO_NAME} is:pr is:closed is:merged linked:issue label:{label} "
                 f"merged:{start_date}..{end_date}")
        params = {
            "q": query,
            "per_page": 100,
            "page": page
        }

        print(f"Fetching page {page} of merged PRs with label '{label}' "
              f"from {start_date} to {end_date}...")
        response = requests.get(BASE_URL, headers=HEADERS, params=params)

        if response.status_code != 200:
            print(f"Failed to fetch merged PRs: {response.status_code} {response.text}")
            break

        results = response.json().get("items", [])
        if not results:  # Break if no more results
            print(f"No more results for interval {start_date} to {end_date}.")
            break

        # Append results to the list
        merged_prs.extend(results)
        page += 1

    return merged_prs


if __name__ == "__main__":
    label_to_filter = "kind/feature"  # Replace with your desired label
    start_date = datetime(2015, 1, 1)  # Replace with your start date
    end_date = datetime(2025, 1, 1)  # Replace with your end date

    current_start_date = start_date
    all_merged_pull_requests = []

    # Loop through the date intervals (per year)
    while current_start_date < end_date:
        current_end_date = datetime(current_start_date.year + 1, 1, 1)  # End of the year
        if current_end_date > end_date:  # Ensure not to exceed the final end date
            current_end_date = end_date

        # Search for PRs in the current year interval
        merged_pull_requests = search_merged_pull_requests_with_label(
            label_to_filter,
            current_start_date.strftime("%Y-%m-%d"),
            current_end_date.strftime("%Y-%m-%d")
        )
        all_merged_pull_requests.extend(merged_pull_requests)
        current_start_date = current_end_date  # Move to the next year

    print(f"Found {len(all_merged_pull_requests)} merged PR(s) with label '{label_to_filter}'.\n")

    # Save the merged pull requests to a JSON file
    output_file = f"merged_pull_requests_with_label_{label_to_filter.replace('/', '_')}.json"
    with open(output_file, "w") as file:
        json.dump(all_merged_pull_requests, file, indent=4)

    print(f"Merged PRs with label '{label_to_filter}' saved to {output_file}.")
```

```python
def get_github_issue_details(issue_url, token=None):
    # Extract owner, repo, and issue number from the URL
    try:
        parts = issue_url.split("/")
        owner = parts[3]
        repo = parts[4]
        issue_number = parts[-1]
    except IndexError:
        return {"error": "Invalid GitHub issue URL."}

    # GitHub API endpoint for issue details
    api_url = f"https://api.github.com/repos/{owner}/{repo}/issues/{issue_number}"

    # Headers for the request
    headers = {
        "Accept": "application/vnd.github+json",
    }
    if token:
        headers["Authorization"] = f"Bearer {token}"

    # Make the GET request
    response = requests.get(api_url, headers=headers)

    # Check for successful response
    if response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Failed to fetch issue details. Status code: {response.status_code}", "details": response.text}

if __name__ == "__main__" :
    # Fetch issue details and save results to a file
    results = []
    for issue_url in issue_urls:
        details = get_github_issue_details(issue_url, token=TOKEN)
        results.append(details)
    
    # Save the results to a JSON file
    output_file = "github_issues_details.json"
    with open(output_file, "w") as file:
        json.dump(results, file, indent=4)
    
    print(f"Details saved to {output_file}")
```

# Get All bug issues

```python
MAX_CONCURRENT_REQUESTS = 10  # Sesuaikan dengan rate limit GitHub

# Header untuk otentikasi
headers = {
    "Authorization": f"Bearer {TOKEN}",
    "Accept": "application/vnd.github.v3+json"
}

def get_total_pages():
    """Mendapatkan total halaman dari header 'Link'"""
    endpoint = f"/repos/{REPO_OWNER}/{REPO_NAME}/issues"
    params = {
        "state": "closed",
        "labels": "kind/bug",
        "per_page": 100,
        "page": 1
    }
    
    with requests.Session() as session:
        session.headers.update(headers)
        response = session.get(f"{BASE_URL}{endpoint}", params=params)
        
        if response.status_code != 200:
            print("Gagal mendapatkan total halaman")
            return 1  # Fallback ke 1 halaman
        
        # Parsing header 'Link' untuk mendapatkan halaman terakhir
        link_header = response.headers.get("Link", "")
        if "rel=\"last\"" in link_header:
            last_page_url = [link for link in link_header.split(", ") if "rel=\"last\"" in link][0]
            last_page = int(last_page_url.split("page=")[-1].split(">")[0])
            return last_page
        else:
            return 1  # Hanya ada 1 halaman

def fetch_page(page: int):
    """Mengambil data dari halaman spesifik dengan retry"""
    endpoint = f"/repos/{REPO_OWNER}/{REPO_NAME}/issues"
    params = {
        "state": "closed",
        "labels": "kind/bug",
        "per_page": 100,
        "page": page
    }
    for _ in range(3):
        with requests.Session() as session:
            session.headers.update(headers)
            response = session.get(f"{BASE_URL}{endpoint}", params=params)
        
            if response.status_code == 200:
                # Filter: Ambil hanya item yang TIDAK memiliki key 'pull_request'
                issues = [
                    issue for issue in response.json() 
                    if "pull_request" not in issue
                ]
                return page, issues

def main():
    print("Memulai pengambilan data...")
    
    # Langkah 1: Dapatkan total halaman
    total_pages = get_total_pages()
    print(f"Total halaman yang akan diambil: {total_pages}")
    
    # Langkah 2: Ambil semua halaman secara paralel
    all_issues = []
    pages = list(range(1, total_pages + 1))
    
    with ThreadPoolExecutor(max_workers=MAX_CONCURRENT_REQUESTS) as executor:
        # Buat task untuk setiap halaman
        future_to_page = {executor.submit(fetch_page, page): page for page in pages}
        
        # Proses hasil segera setelah selesai
        for future in as_completed(future_to_page):
            page = future_to_page[future]
            try:
                p, issues = future.result()
                print(f"Berhasil mengambil {len(issues)} issue dari halaman {p}")
                all_issues.extend(issues)
            except Exception as e:
                print(f"Error pada halaman {page}: {str(e)}")
    
    # Langkah 3: Simpan hasil
    with open("github_issues_3.json", "w") as f:
        json.dump(all_issues, f, indent=4)
    
    print(f"\nTotal issue yang berhasil diambil: {len(all_issues)}")
    print("Data tersimpan di github_issues_3.json")

if __name__ == "__main__":
    main()
```
