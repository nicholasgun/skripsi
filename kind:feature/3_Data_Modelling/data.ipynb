```python
import pandas as pd
from collections import Counter
```

```python
data_path = '../../Data/Preprocessed Data/kind:feature/cleaned_data_with_changed_files.json'
df = pd.read_json(data_path)
```

```python
df.head()
```

```python
df = df[['all_text','labels']]
```

```python
df.head()
```

```python
# sort df by all_text count asc
df['all_text_count'] = df['all_text'].apply(lambda x: len(x.split()))
df = df.sort_values('all_text_count', ascending=True)
```

```python
df
```

```python
# check for all_text data that starts with nan
df[df['all_text'].apply(lambda x: x.startswith('nan'))]
```

```python
# remove all_text data that starts with nan
df = df[~df['all_text'].apply(lambda x: x.startswith('nan'))]
```

```python
data = df['all_text']
labels = df['labels'].apply(lambda x : x if isinstance(x, list) else [])
```

```python
# check if there's null data in values
labels.isnull().values.any()
```

```python
label_distribution = Counter([label for labels in labels for label in labels])

print('Label Distribution (Exploded labels):')
for label, count in sorted(label_distribution.items(), key=lambda x: x[1], reverse=True):
    print(f'{label}: {count}')
```

```python
# Get labels with count >= 5
frequent_labels = [label for label, count in label_distribution.items() if count >= 5]
print(f"Number of labels with at least 5 occurrences: {len(frequent_labels)}")
print(f"Removed {len(label_distribution) - len(frequent_labels)} infrequent labels")

# Filter each row's labels to keep only frequent labels
filtered_labels = []
for label_list in labels:
    filtered_label_list = [label for label in label_list if label in frequent_labels]
    filtered_labels.append(filtered_label_list)

# Create a new dataset with filtered labels
# Assuming your dataframe has the original labels and texts
df_filtered = pd.DataFrame({
    'text': data,  # assuming you have a texts variable
    'labels': filtered_labels
})

# Remove rows that now have empty label lists (all labels were infrequent)
df_filtered = df_filtered[df_filtered['labels'].apply(len) > 0]

print(f"Original dataset size: {len(df)}")
print(f"Filtered dataset size: {len(df_filtered)}")

# Check the new label distribution
new_label_distribution = Counter([label for labels in filtered_labels for label in labels])
print('\nLabel Distribution after filtering:')
for label, count in sorted(new_label_distribution.items(), key=lambda x: x[1], reverse=True):
    print(f'{label}: {count}')
```

```python
# label length distribution
label_length_distribution = Counter([len(labels) for labels in labels])

print('Label count per row distribution:')
for label in sorted(label_length_distribution.keys()):
    print(f'Label: {label}, count: {label_length_distribution[label]}')
```

```python
# get the distribution of label combinations if there are multiple labels
label_combinations = Counter([tuple(labels) for labels in labels])

# Or for descending order by count
for i, (labels, count) in enumerate(sorted(label_combinations.items(), key=lambda x: x[1], reverse=True)):
    print(f'{i}. Label combination: {labels}, count: {count}')   
```
